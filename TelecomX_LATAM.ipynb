```python
import pandas as pd
```

```python
telecomX_data = pd.read_json('https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json')
```

#ðŸ“Š Carga y anÃ¡lisis

```python
# Normalizando las columnas 'customer', 'phone', 'internet', y 'account'
customer_data = pd.json_normalize(telecomX_data['customer'])
phone_data = pd.json_normalize(telecomX_data['phone'])
internet_data = pd.json_normalize(telecomX_data['internet'])
account_data = pd.json_normalize(telecomX_data['account'])

# Concatenando los dataframes normalizados con el dataframe original
telecomX_normalized = pd.concat([telecomX_data[['customerID', 'Churn']],
                                 customer_data,
                                 phone_data,
                                 internet_data,
                                 account_data], axis=1)
```

```python
# Convertir columnas 'Yes'/'No' a boolean (incluyendo 'Churn')
yes_no_cols = ['Churn', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity',
               'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
               'StreamingMovies', 'PaperlessBilling']

for col in yes_no_cols:
    telecomX_normalized[col] = telecomX_normalized[col].map({'Yes': True, 'No': False}).astype(bool)

# Manejar 'No phone service' y 'No internet service' en MultipleLines y otras columnas relacionadas con internet
# Re-aplicando el mapeo y luego astype(bool) para asegurar el manejo correcto de "No phone service" y "No internet service"
telecomX_normalized['MultipleLines'] = telecomX_normalized['MultipleLines'].map({'Yes': True, 'No': False, 'No phone service': False}).astype(bool)
for col in ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
               'StreamingTV', 'StreamingMovies']:
     telecomX_normalized[col] = telecomX_normalized[col].map({'Yes': True, 'No': False, 'No internet service': False}).astype(bool)

# Convertir 'Charges.Total' a numÃ©rico, forzando errores
telecomX_normalized['Charges.Total'] = pd.to_numeric(telecomX_normalized['Charges.Total'], errors='coerce')

```
